```python
# Create an empty list to store names
namelist = []

# Define a function to store a name in the list
def storeName():
    # Prompt the user to enter a name and store it in the 'name' variable
    name = input("Enter the name")
    # Remove leading and trailing whitespace and convert the name to title case
    name = name.strip().title()
    # Append the formatted name to the 'namelist' list
    namelist.append(name)
    # Return the formatted name (not used in the current code)
    return name

# Define a function to list all the names in 'namelist'
def listNames():
    # Print a line of asterisks for visual separation
    print("*"*30)
    print("Names in the list")
    print("*"*30)
    # Iterate through each name in 'namelist' and print it
    for name in namelist:
        print(name)
    # Print another line of asterisks for visual separation
    print("*"*30)

# Define a function to search for a name in 'namelist'
def searchName():
    # Splitting 'search' variable (which is undefined) doesn't make sense
    # It's better to get the search query as input from the user
    search = input("Enter the name to search: ").strip().title()
    found = False
    # Iterate through each name in 'namelist' and check if it matches the search query
    for name in namelist:
        if name == search:
            found = True
            break
            # If a match is found, print a message
        if found == True :
            print("Name exists in the list")
        else :
            print("Name doesnt exist")
# Create an infinite loop for the menu
while True:
    print("*"*30)
    print("1. Enter the names")
    print("2. List the names")
    print("3. Search for a Name")
    print("4. Exit")
    print("*"*30)

    # Prompt the user for their choice and convert it to an integer
    choice = int(input("Enter your Choice ?"))
    print("You have entered choice", choice)
    
    # Check the user's choice and perform the corresponding action
    if choice == 1:
        # Call the 'storeName' function to add a name to the list
        name = storeName()
        print("Name {} added successfully".format(name))
    elif choice == 2:
        # Call the 'listNames' function to list all the names
        listNames()
    elif choice == 3:
        # Call the 'searchName' function to search for a name
        searchName()
    elif choice == 4:
        # Exit the program
        break
       
    else:
        # If an invalid choice is entered, print an error message
        print("Invalid Syntax")
```

```python
# Library Management System

# Initialize an empty list to store library books
library_books = []

# Function to add a new book to the library
def add_book():
    title = input("Enter book title: ")
    author = input("Enter author's name: ")
    isbn = input("Enter ISBN number: ")
    library_books.append({"Title": title, "Author": author, "ISBN": isbn, "Status": "Available"})

# Function to check out a book from the library
def check_out_book():
    isbn = input("Enter ISBN of the book to check out: ")
    for book in library_books:
        if book["ISBN"] == isbn:
            if book["Status"] == "Available":
                book["Status"] = "Checked Out"
                print(f"Book '{book['Title']}' has been checked out.")
            else:
                print(f"Book '{book['Title']}' is already checked out.")
            return
    print("Book not found in the library.")

# Function to display the list of available books in the library
def display_available_books():
    available_books = [book for book in library_books if book["Status"] == "Available"]
    if available_books:
        print("Available Books:")
        for book in available_books:
            print(f"Title: {book['Title']}, Author: {book['Author']}, ISBN: {book['ISBN']}")
    else:
        print("No books are available in the library.")

# Function to generate a report of all books in the library
def generate_book_report():
    print("Library Book Report:")
    for book in library_books:
        print(f"Title: {book['Title']}, Author: {book['Author']}, ISBN: {book['ISBN']}, Status: {book['Status']}")

# Main program loop
while True:
    print("-" * 30)
    print("Library Management System")
    print("1. Add Book")
    print("2. Check Out Book")
    print("3. Display Available Books")
    print("4. Generate Book Report")
    print("5. Exit")
    
    choice = int(input("Enter your choice: "))
    
    if choice == 1:
        add_book()
    elif choice == 2:
        check_out_book()
    elif choice == 3:
        display_available_books()
    elif choice == 4:
        generate_book_report()
    elif choice == 5:
        print("End of program.")
        break
    else:
        print("Invalid choice. Please select a valid option.")
```
